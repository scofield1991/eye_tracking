```python
from sklearn import cross_validation,metrics
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn 

%matplotlib inline
```

```python
df_gen  = pd.read_csv('deeplearning.csv', chunksize=1000000,sep = ";")
df=df_gen.next()
df.head()
```

```python
gr=df.groupby('uid')
plt.figure(figsize=(15,8))
for n_col,col in enumerate(['left_pd','right_pd','left_px','right_px','left_x','fix']):
    plt.subplot(3,2,n_col+1)
    for i in range(1591,1594):
        # print gr.mean()[i]
        x = gr.trial_time.apply(lambda x: x.values).iloc[i]
        y = gr[col].      apply(lambda x: x.values)[i]
        plt.plot(x,y)
        plt.title(col)
```
